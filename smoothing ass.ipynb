```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read and preprocess the image
image = cv2.imread('sun.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Define kernel sizes and parameters
kernel_size = 7
sigma = 0.05 * ((kernel_size - 1) * 0.5 - 1) + 2.35
```

```python
# Apply filtering techniques
avg_kernel = np.ones((kernel_size, kernel_size), np.float32) / (kernel_size ** 2)
avg_image = cv2.filter2D(gray, -1, avg_kernel)
```

```python
gauss_kernel = cv2.getGaussianKernel(kernel_size, sigma)
gauss_image = cv2.filter2D(gray, -1, gauss_kernel)
```

```python
median_image = cv2.medianBlur(gray, kernel_size)

bilateral_image = cv2.bilateralFilter(gray, d=9, sigmaColor=95, sigmaSpace=75)
```

```python
plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 1)
plt.imshow(avg_image, cmap='gray')
plt.title('Averaging')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(bilateral_image, cmap='gray')
plt.title('Bilateral Filtering')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(gauss_image, cmap='gray')
plt.title('Gaussian Filtering')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(median_image, cmap='gray')
plt.title('Median (Kernel Size 15)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
