```python
img= cv2.imread('lenna.png')
```

```python
img = cv2.imread('lenna.png')
```

```python
r, g, b = cv2.split(img)
```

```python
plt.figure(figsize=(16,9))
plt.title('Color image histogram')
plt.subplot(1,3,1)
plt.xticks(range(0,256,10))
plt.hist(b.flatten(), bins = range(0,256), color = 'b');

plt.subplot(1,3,2)
plt.xticks(range(0,256,10))
plt.hist(g.flatten(), bins = range(0,256), color = 'g');

plt.subplot(1,3,3)
plt.xticks(range(0,256,10))
plt.hist(r.flatten(), bins = range(0,256), color = 'r');

plt.tight_layout()
```

```python
import cv2

image = cv2.imread('lenna.png')


new_width , new_height = 200,512

streached_image = cv2.resize(image,dsize = (new_width , new_height),interpolation=cv2.INTER_CUBIC)

cv2.imshow('Stretched Image',streached_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
## histogram equalization
import numpy as np

img = cv2.imread('lenna.png')
gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
equ = cv2.equalizeHist(gray)


cv2.imshow('image',equ)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
