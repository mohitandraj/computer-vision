```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image = cv2.imread('sachin.jpg')
```

<h5>1. Size and shape of the image</h5>

```python
height, width, channels = image.shape
```

```python
height, width, channels
```

<h5>2. The color space model</h5>

```python
if image is not None:
    print("Color Space: BGR") 
else:
    print("Color Space: Unknown")
```

<h5>3. Display using cv and matplotlib</h5>

```python
# Display using OpenCV
cv2.imshow('Image', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Display using Matplotlib
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Sachin Image")
plt.show()
```

<h5> Show the image in the matrix of 5 x 5 with following colormaps:
1. coolwarm
2. rainbow
3. tab10
4. seismic
5. twilight
6. turbo
7. BuGn</h5>

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 7, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2GRAY), cmap='coolwarm')
plt.title('coolwarm')
plt.axis('off')

plt.subplot(1, 7, 2)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2GRAY), cmap='rainbow')
plt.title('rainbow')
plt.axis('off')

plt.subplot(1, 7, 3)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2GRAY), cmap='tab10')
plt.title('tab10')
plt.axis('off')

plt.subplot(1, 7, 4)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2GRAY), cmap='seismic')
plt.title('seismic')
plt.axis('off')

plt.subplot(1, 7, 5)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2GRAY), cmap='twilight')
plt.title('twilight')
plt.axis('off')

plt.subplot(1, 7, 6)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2GRAY), cmap='turbo')
plt.title('turbo')
plt.axis('off')

plt.subplot(1, 7, 7)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2GRAY), cmap='BuGn')
plt.title('BuGn')
plt.axis('off')

plt.tight_layout()
plt.show()
```

<h5>Change the pixels from (100,150) to (200,350) to while color.</h5>

```python
image[100:200, 150:350] = [255, 255, 255]  
cv2.imshow('Modified Image', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

<h4>Video</h4>

```python
vid=cv2.VideoCapture('nature.mp4')
```

```python
width = int(vid.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(vid.get(cv2.CAP_PROP_FRAME_HEIGHT))
frames = int(vid.get(cv2.CAP_PROP_FRAME_COUNT))
fps= int(vid.get(cv2.CAP_PROP_FPS))
```

```python
width, height, frames, fps
```

```python
video = cv2.VideoCapture('nature.mp4')
if video.isOpened():
    while True:
        ret, frame = video.read()
        
        if ret:
            cv2.imshow('Video Frame', frame)
                       
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break
        else:
            break  
else:
    print("Failed to open video.")

video.release()
cv2.destroyAllWindows()
```

```python
video = cv2.VideoCapture('nature.mp4')
fps = video.get(cv2.CAP_PROP_FPS)
frame_index = int(fps * 50)
video.set(cv2.CAP_PROP_POS_FRAMES, frame_index)
ret, frame = video.read()
cv2.imwrite('frame_at_50s.jpg', frame)
video.release()
cv2.destroyAllWindows()
```
