```python
import cv2
import numpy as np

image = cv2.imread('lenna.png')

kernal = np.ones((5,5),np.float32) / 25

smoothed_image = cv2.filter2D(image,-1,kernal)

cv2.imshow('Original',image)
cv2.imshow('Smoothed',smoothed_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
final_image= np.vstack([image, smoothed_image])
```

```python
cv2.imshow("Final",final_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('lenna.png')
gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
 kernal_size = 15
median = cv2.medianBlur(gray,kernal_size)

plt.figure(figsize=(15,10))
plt.subplot(1,2,1)
plt.imshow(gray,cmap = 'gray')
plt.title('Original img')
plt.subplot(1,2,2)
plt.imshow(median,cmap = 'gray')
plt.title('Median BLURRED img')
plt.show()
```

```python
median3 = cv2.medianBlur(gray, 3)
median5 = cv2.medianBlur(gray, 5)
median7 = cv2.medianBlur(gray, 7)
median9 = cv2.medianBlur(gray, 9)
median11 = cv2.medianBlur(gray, 11)
median13 = cv2.medianBlur(gray, 13)
```
