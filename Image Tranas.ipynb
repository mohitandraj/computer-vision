```python
image= cv2.imread("lenna.png")

rotated_image = cv2.rotate(image, cv2.ROTATE_90_COUNTERCLOCKWISE)
cv2.imshow('ROTATED IMAGE', rotated_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
image= cv2.imread("lenna.png")

center= (image.shape[1] // 2, image.shape[0] //2)
angle= 80

rotation_matrix = cv2.getRotationMatrix2D(center, angle , 1.0)

rotated_image= cv2.warpAffine(image, rotation_matrix, (image.shape[1], image.shape[0]))

np_image = np.ones([image.shape[0],image.shape[1],3], dtype=np.uint8)+254

for i in range(512):
    for j in range(512):
        for k in range(3):
            if rotated_image[i][j][k] !=0:
                np_image[i][j][k]= rotated_image[i][j][k]
                
cv2.imshow("rotated image",rotated_image)
cv2.imshow("new image", np_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
image= cv2.imread('lenna.png')

resized_image = cv2.resize(image,(100,200))

cv2.imshow("resized image", resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
resized_image = cv2.resize(image,None, fx=1.5, fy=1.5)

cv2.imshow("resized image", resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
resized_image = cv2.resize(image,None, fx=2.5, fy=2.5)

cv2.imshow("resized image", resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
resized_image = cv2.resize(image,None, fx=1.5, fy=1.5,
                          interpolation = cv2.INTER_NEAREST)

cv2.imshow("resized image", resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
##### image resizing using skimage
```

```python
from skimage.transform import resize
from skimage import io
import matplotlib.pyplot as plt
image = io.imread('lenna.png')

resized_image = resize(image,(1000,1200),anti_aliasing=True)

resized_image = resize(image,(1300,1300),anti_aliasing=True, order =3)

plt.subplot(1,2,1)
plt.imshow(image)
plt.subplot(1,2,2)
plt.imshow(resized_image)

```

```python
### image flipping
```

```python
image= cv2.imread('lenna.png')
```

```python
flipped= np.flip(image)
```

```python
cv2.imshow("flipped image",flipped)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
flipped_horizontal = cv2.flip(image,0)
flipped_vertical = cv2.flip(image,1)
flipped_diagonal = cv2.flip(image,-1)

cv2.imshow('Original',image)
cv2.imshow('Horizontal',flipped_horizontal)
cv2.imshow('Vertical',flipped_vertical)
cv2.imshow('Diagonal',flipped_diagonal)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from PIL import Image

image = Image.open('lenna.png')

flipped_image = image.transpose(Image.FLIP_LEFT_RIGHT)
flipped_image.show()

flipped_image = image.transpose(Image.FLIP_TOP_BOTTOM)
flipped_image.show()

flipped_image = image.transpose(Image.TRANSPOSE)
flipped_image.show()
```
